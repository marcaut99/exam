```python
data = dataset.language.value_counts()
labels = dataset.language.unique()
```

```python
counts_less_than_something = data<10
```

```python
new_data = data[~counts_less_than_something]
```

```python
new_data["Other languages"] = data[counts_less_than_something].sum()
```

```python
new_data
```

```python
plt.pie(new_data,
        labels=new_data.index,
        autopct = '%.1f%%'
        )
plt.title("Languages", fontsize = 20)
plt.show()
```

```python
plt.pie(data,
        labels=labels,
        autopct = '%.1f%%'
        )
plt.title("Languages", fontsize = 20)
plt.show()
```
